
Exercise 1.2: Long Run Divisibilities

```python
import random as rnd
import matplotlib.pyplot as plt

n = 1000
by7 = 0
by7OverTime = []
by7and3 = 0
by7and3OverTime = []
for i in range(n):
    num = (int(50*rnd.random()))
    if (num % 7) == 0:
        by7 += 1
        if (num % 3) == 0:
            by7and3 += 1
    by7OverTime.append(by7/(i+1))
    by7and3OverTime.append(by7and3/(i+1))
print(f"The observed frequency of event 1 occuring in {n} draws is {by7}")
print(f"The observed frequency of event 2 occuring in {n} draws is {by7and3}")

expBy7 = 0
expBy7and3 = 0

for i in range(50):
    if (i % 7) == 0:
        expBy7 += 1
        if (i % 3) == 0:
            expBy7and3 += 1

print(f"The expected frequency of event 1 occuring in {n} draws is {(expBy7 / 49) * n} or a proportion of {expBy7 / 49}")
print(f"The expected frequency of event 2 occuring in {n} draws is {(expBy7and3 / 49) * n} or a proportion of {expBy7and3 / 49}")

plt.plot(range(1, n+1), by7OverTime)
plt.xlabel("Number of Samples")
plt.ylabel("Observed frequency of divisibility")
plt.title("Probability is divisible by 7")
plt.show()

plt.plot(range(1, n+1), by7and3OverTime)
plt.xlabel("Number of Samples")
plt.ylabel("Observed frequency of divisibility")
plt.title("Probability is divisible by 7 and 3")
plt.show()
```
