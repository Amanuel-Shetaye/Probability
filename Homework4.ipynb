```python
import numpy as np
import random
import matplotlib.pyplot as plt

alpha = 2  
lambdas = 1  

# Number of samples to simulate
num_samples = 10000
numbers = []
#N random samples
for i in range(num_samples):
   num = random.random()
   numbers.append(num)
numpyArray = np.array(numbers)

#T = (-log(U) / lambda)^(1/alpha)
T = (-np.log(numpyArray) / lambdas)**(1/alpha)

plt.hist(T, bins=50, color='blue',edgecolor='black', density= True,label="Simulated Weibull")

t_values = np.linspace(0, np.max(T), 1000)
weibull_pdf = lambdas * alpha * t_values**(alpha - 1) * np.exp(-lambdas * t_values**alpha)

plt.plot(t_values, weibull_pdf, lw=2, label=f"Real Weibull PDF\n(λ = {lambdas}, α = {alpha})")

plt.title(f"Simulated Weibull Distribution λ = {lambdas} and α = {alpha}")
plt.xlabel("Value of T")
plt.ylabel("Density")
plt.legend()

plt.show()
```
