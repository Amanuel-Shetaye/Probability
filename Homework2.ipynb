Excercise 3.1: Simulating Discrete Distribution

```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt

def genBer(vec):
    q = np.cumsum(vec)
    q = np.append(0,q)
    num = rnd.random()
    for i in range(len(q) - 1):
        if num > q[i] and num < q[i+1]:
            return i
prtLis = []
for i in range(10):
    prtLis.append((genBer([.1,.5,.4])))
print(prtLis)
prtLis = []
for i in range(10):
    prtLis.append((genBer([.4,.2,.1,.3])))
print(prtLis)
prtLis = []
for i in range(10):
    prtLis.append((genBer([.3,.3,.1,.1,.1,.1])))
print(prtLis)
```

Excercise 3.2: Checking Excercise 3.1

```python
for count in [5, 10, 50, 100, 500, 1000]:
    total0 = 0
    total1 = 0
    total2 =0
    for i in range(count):
        draw = genBer([0.6,0.1,0.3])
        if draw == 0:
            total0 += 1
        elif draw == 1:
            total1 += 1
        else:
            total2+= 1

    print(f"In {count} draws the average amount of 0's was {total0/count}, \nthe average amount of 1's was {total1/count}, \nand the average amount of 2's was {total2/count}.")



total0 = 0
prop0 = []
total1 = 0
prop1 = []
total2 =0
prop2 = []

for i in range(1000):
    draw = genBer([0.6,0.1,0.3])
    if draw == 0:
        total0 += 1
    elif draw == 1:
        total1 += 1
    else:
        total2+= 1
    prop0.append(total0/(i+1))
    prop1.append(total1/(i+1))
    prop2.append(total2/(i+1))

plt.plot(range(1, 1001), prop0)
plt.xlabel("Number of Samples")
plt.ylabel("Average Proportion of 0's drawn")
plt.title("Probability of drawing 0")
plt.show()
plt.plot(range(1, 1001), prop1)
plt.xlabel("Number of Samples")
plt.ylabel("Average Proportion of 1's drawn")
plt.title("Probability of Drawing 1")
plt.show()
plt.plot(range(1, 1001), prop2)
plt.xlabel("Number of Samples")
plt.ylabel("Average Proportion of 2's drawn")
plt.title("Probability of drawing 2")
plt.show()
```

It appears that the function I made is working becuase as the number of samples increase the average number of any of the indexes begins to approach the desired probabiltiy.

Excercise 3.7 Monte Carlo Convergence Rate

```python
def f(x):
    return ((0.5)*(x**3)-(x**2)+1)

under = 0
propUnder = []
error = []
n = 1000
for i in range(n):
    x = 2*rnd.random()
    y = rnd.random()
    if f(x) > y:
        under += 1
    tempPropUnder = (under/(i+1))
    estimate = tempPropUnder * 2
    propUnder.append(estimate)
    trueValue = 4/3
    error.append(abs((estimate-(trueValue))/trueValue))




plt.plot(range(1, n+1), error)
plt.xlabel("Number of Samples")
plt.ylabel("Error")
plt.title("Error of Monte Carlo Estimation")
plt.show()

error = np.array(error)
samples = np.arange(1, n + 1)

log_samples = np.log(samples)
log_error = np.log(error + 1e-10)

plt.plot(log_samples, log_error)
plt.xlabel("Log Number of Samples")
plt.ylabel("Log Error")
plt.title("Log Error of Monte Carlo Estimation")
plt.show()

```

As the number of trials increases the error rapidly approaches 0. This makes sense because approximations are more accurate with more trials.
